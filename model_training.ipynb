```python
from transformers import AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained("bert-base-cased")
text="I was so not happy with the Barbie Movie."
tokens=tokenizer.tokenize(text)
print("Tokens:",tokens)
input_ids=tokenizer.convert_tokens_to_ids(tokens)
print("INPUT IDS:",input_ids)
encoded_input=tokenizer(text)
print("encoded_input:",encoded_input)
decoded_output=tokenizer.decode(input_ids)
print("decoded output:",decoded_output)

```

```python
from datasets import load_dataset
dataset=load_dataset('imdb')
```

```python
dataset
```

```python
from transformers import AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained("bert-base-cased")
def tokenize_function(examples):
  return tokenizer(examples['text'],padding="max_length",truncation=True)
tokenized_datasets=dataset.map(tokenize_function, batched=True)
```

```python
from transformers import TrainingArguments
training_args=TrainingArguments(
    output_dir='./results',
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_eval_batch_size=16,
    per_device_train_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,

)
training_args
```

```python
from transformers import AutoModelForSequenceClassification, Trainer

model = AutoModelForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=2
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['test']
)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"  # MUST be before creating training_args

from transformers import AutoModelForSequenceClassification, Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    report_to=[],        # prevents logging to wandb
    logging_dir="./logs" # optional: local logs
)

model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"]
)

trainer.train()
```

```python
results=trainer.evaluate()
print(results)
```
